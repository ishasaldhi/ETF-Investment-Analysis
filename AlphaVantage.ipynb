## Starter Code to Generate JSON files for ETF Data

```python
import requests
import json
# Import Alpha Vantage API key
from config import api_key

from pymongo import MongoClient
from pprint import pprint
```

```python
# Specify the stock symbol 
symbols1 = ['SPY', 'QQQ', 'IWM', 'EEM', 'AGG']
symbols2 = ['VTI', 'GLD', 'SLV', 'DBC', 'VNQ']
# Define the company data variable
company_data = {}
```

```python
# Send an HTTP GET request to the API
#response = requests.get(url)

# Check if the request was successful 
for symbol in symbols1:
    url1 = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol={symbol}&apikey={api_key}'
    response = requests.get(url1)
    
    if response.status_code == 200:
           # Parse the JSON response
        data = response.json()

        # Store the data for the current company in the dictionary
        company_data[symbol] = data
    else:
        # Handle the case where the request was not successful for a company
        print(f"Request failed for {symbol} with status code {response.status_code}")
        
        
        
        
output_file = 'alpha_vantage_data.json'
with open(output_file, 'w') as json_file:
    json.dump(company_data, json_file, indent=4)

print(f"Data saved to '{output_file}'")


```

```python
import time

symbols1 = ['SPY', 'QQQ', 'IWM', 'EEM', 'AGG']


company_details = []

for i in range(len(symbols1)):
    # API URL
    url = f'https://www.alphavantage.co/query?function=TIME_SERIES_WEEKLY_ADJUSTED&symbol={symbols1[i]}&apikey={api_key}'
    
    # Make the API request
    response = requests.get(url)
    print(response.status_code)
    # Extract the json result
    extract_json = response.json()
    
    # Add the returned json to the end of the list
    company_details.append(extract_json)
    time.sleep(2)
    
    

# Print the length of the results to ensure it matches the cave_ids list length
print(f'Length of company details: {len(company_details)}')
```
